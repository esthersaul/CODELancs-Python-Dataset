```python
import seaborn as sns
```

```python
fig, axes = plt.subplots(figsize=(15, 10))
aircraft_data_['Aircraft_Phase'].value_counts().plot(kind='pie')
fig.suptitle('Aircraft Phase When Incident Occured')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(20, 10))
ax = sns.barplot(aircraft_data_, x="Incident_Month", y="Fatalities", estimator="sum", errorbar=None, palette='rocket')
ax.bar_label(ax.containers[0], fontsize=10);
fig.suptitle('Fataliites vs Month')

fig, axes = plt.subplots(figsize=(20, 10))
ax = sns.barplot(aircraft_data_, x="Incident_Day", y="Fatalities", estimator="sum", errorbar=None, palette='rocket')
ax.bar_label(ax.containers[0], fontsize=4);
fig.suptitle('Fataliites vs Day')

```

```python
fig, axes = plt.subplots(figsize=(20, 10))
aircraft_data_['Incident_Month'].value_counts().plot(kind='bar' , color='lightgreen')
ax.bar_label(ax.containers[0], fontsize=4);
fig.suptitle('Number of Incidents Per Month')

fig, axes = plt.subplots(figsize=(20, 10))
aircraft_data_['Incident_Day'].value_counts().plot(kind='bar', color='plum')
ax.bar_label(ax.containers[0], fontsize=4);
fig.suptitle('Number of Incidents Per Day of the Week')

fig, axes = plt.subplots(figsize=(20, 10))
aircraft_data_['Incident_Year'].value_counts().plot(kind='bar', color='orange')
ax.bar_label(ax.containers[0], fontsize=0.5);
fig.suptitle('Number of Incidents Per Year')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
aircraft_data_['Incident_Month'].plot.hist(ax=axes[0], bins=50, alpha=0.5, color='hotpink')
axes[0].set_title('Month')

aircraft_data_['Incident_Year'].plot.hist(ax=axes[1], bins=50, alpha=.5, color='blue')
axes[1].set_title('Year')

axes[1].set_ylabel('')
```

```python
fig, axes = plt.subplots(figsize=(15, 10))
aircraft_data_['Aircaft_Nature'].value_counts().plot(kind='pie', fontsize=8)
fig.suptitle('Aircraft Nature')
plt.tight_layout()
```
